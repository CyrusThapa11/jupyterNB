```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df = pd.read_csv('movie_metadata.csv')
df.head()
```

```python
print(df.columns)
print(type(df.columns))
```

```python
movieTitle = list( df.get('movie_title') )
print(type(movieTitle))
print(movieTitle)
```

```python
score = list( df.get('imdb_score') )
print(score)
```

```python
# df.shape :
```

```python
for t in movieTitle:
    lenn = len(t)
    print(lenn , t )
```

```python
freq_title = {}
for t in movieTitle :
    lenn = len(t)
        
    try: # here we can also use if else block !!
        freq_title[lenn] += 1
    except:
        freq_title[lenn] = 1

print(freq_title)
```

```python
X = np.array(list(freq_title.keys()))
Y = np.array(list(freq_title.values()))
```

```python
plt.scatter(X, Y)
plt.xlabel('movie title len :')
plt.xlabel('no of movies :')
plt.show() # looks like normal dist. !
```
