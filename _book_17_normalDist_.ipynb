```python
u = 5
sigma = 2
```

```python
vals = u + sigma*np.random.randn(100)
```

```python
print(vals)
```

```python
import matplotlib.pyplot as plt  
```

```python
plt.hist( vals, 10  ) # 50 is the thickness here !
plt.show()
```

```python
# multivarient ND : we need mean vector and the variable vector !
```

```python
mean = np.array([ 0.0 , 0.0 ]  )
cov = np.array( [ [1 , 0] ,[ 0 , 1 ] ]  )
dist = np.random.multivariate_normal( mean , cov , 500 )
print(dist )
```

```python
print(dist.shape)
```

```python
plt.scatter ( dist[ : , 0 ]  , dist [ : , 1  ] )
plt.show() # here is mean is 0 of corelation ! 
```
