```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# DUMMY DATA  :
x = np.arange(10)
y1 = x**2
y2 = 2*x + 4
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1) # x will go on x axis and y1 on y axis !
plt.show()
```

```python
plt.style.use('bmh')
plt.plot(x,y1 , color='red',label = 'apple' , marker='X')  
plt.plot(x , y2 , color='green' , label = 'kiwi' , linestyle = 'dashed')
plt.xlabel('time')
plt.ylabel('price ')
plt.title('prices of fruites over time !')
plt.legend() # this needs to be written for the label of lines !! apple and kiwi 
plt.show()
```

```python
# PLOTTING A HISTOGRAM !!
Xsn = np.random.randn(100) # generate smaple data of std normal dist !
sigma = 5
u = 60
X = np.round(Xsn*sigma + u)
X2 = np.round(Xsn*8 + 40) # higher SD higher the overlap !
print(X)
```

```python
plt.hist(X,alpha = 0.7 , label="maths")

plt.hist(X2,alpha = 0.8,label="physics ")
plt.ylabel('Probability / frequency count of students ')
plt.xlabel('marks range')

plt.show()
```

```python
# a little project !
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./movie_metadata.csv')
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles = df.get('movie_title')
print(type(titles))
# typecast to python list !
titles_2 = list(df.get('movie_title'))
print(titles_2[:10])
```

```python
freq_titles = {}

for title in titles:
    lenn = len(title)
    
    if freq_titles.get(lenn) is None:
        freq_titles[lenn] = 1
    else:
        freq_titles[lenn] += 1
```

```python
print(freq_titles)
```

```python
# now we need to make a graph !
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
print(X , Y)
```

```python
plt.scatter(X,Y)
plt.xlabel('length of movie title ')
plt.ylabel('freq count ')
plt.title('movie data :')
plt.show()
```
