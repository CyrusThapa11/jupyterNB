```python
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#grey_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image) # this is original image !
```

```python
plt.axis('off')
plt.imshow(rgb_image)
```
